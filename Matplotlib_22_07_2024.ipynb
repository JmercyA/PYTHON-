```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
year = np.arange(2005,2020)
len(year)
```

```python
Price = np.linspace(50,150,15)
Price.round(2)
```

```python
units_sold = np.linspace(10,30,15)
```

#1. Line Graph/Line Plot

```python
# Used when period(Time element) is involved.
# produces trends(upwards,downwards,cyclic)
fig = plt.figure(figsize=(10,5))
plt.plot(year,Price, color = 'red')
plt.plot(year,units_sold, color = 'green')
plt.title('price per year analysis')
plt.xlabel('year')
plt.ylabel('price')
plt.show()
```

```python
data1 = np.array([20,30,50,80])
fig = plt.figure(figsize=(8,4))
plt.plot(data1,linestyle ='dotted');
plt.title('grades')
plt.xlabel('period')
plt.ylabel("marks");
```

2. Subplot

```python
# used when there are many visualizations that should be presesntes at the same place
```

```python
fig, ax = plt.subplots(2, 2) #display 1 row and 2 columns
ax[0,0].plot(data1,color='red');

ax[0,1].plot(year,Price,color='green')
ax[0,1].plot(year,units_sold, color = 'blue')

ax[1,0].plot(year,Price,color= 'purple')
ax[1,1].plot(year,Price,color = 'k');
```
