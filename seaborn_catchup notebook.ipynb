Seaborn

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print(sns.get_dataset_names(), end= " ")
```

```python
#load the dataset
df= sns.load_dataset('tips')
df.head()
```

```python
sns.scatterplot(x = "total_bill", y = "tip", data=df)
plt.title("bill vs tip")
plt.show()
```

```python
#check r/ship btn bill, tips and gender
sns.scatterplot(x=df["total_bill"], y= df['tip'], hue = df["smoker"] )
plt.show()
```

```python
smoker_count = df["smoker"].value_counts()
smoker_count
```

```python
#visualize the above
smoker_count.plot(kind="bar")
```

```python
df.describe()
```

```python
sns.boxplot(df)
```

```python
#box plot on tip
sns.boxplot(y='tip', data = df);
```

```python
#box plot using numerical and categorical variable
sns.boxplot(y='tip', x="sex", data = df);
```

```python
smoker_count = df["smoker"].value_counts()
smoker_count
```

```python
smoker_sex = pd.crosstab(df["smoker"], df["sex"])
smoker_sex
```

```python
#visualize smoker variable
sns.barplot(smoker_sex, errorbar= None)
plt.ylabel("smoker")
```

```python
sns.countplot(x='smoker', hue='sex', data=df, order = ["No", 'Yes'])
```

```python
#creting subplot for numerical columns
num_df = df[['total_bill', 'tip', 'size']]

num_df.head()
```

```python
#Get correlation coef
corr =num_df .corr()
corr
```

```python
#create heatmap
sns.heatmap(corr, annot= True, fmt='.2f', cmap = "icefire")
```

```python
#check tatal bill distribution
sns.histplot(df['total_bill'], kde=True)
```

```python
#check tatal bill distribution with bin argument
sns.histplot(df['total_bill'], bins = 5, kde=True)
```

```python
sns.kdeplot(df['total_bill'])
```

```python
#violin plot
sns.violinplot(df['total_bill']);
```
